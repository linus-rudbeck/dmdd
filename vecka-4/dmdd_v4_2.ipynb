# Hämta data från MySQL

Här hämtar vi data från en MySQL-databas, städar den och sparar till en CSV.

**Steg 1: importera nödvändiga bibliotek**

- `pandas` för att jobba med data frames
- `mysql.connector` för att ansluta till en mysql-databas

```python
import pandas as pd
import mysql.connector
```

**Steg 2: Skapa en anslutning**

```python
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="", # mac: "root"
    database="northwind"
)

connection
```

**Steg 3: Hämta data**

```python
query = "SELECT * FROM customers"

cursor = connection.cursor()
cursor.execute(query)

data = cursor.fetchall()

data
```

**Steg 4: Spara data som DataFrame**

```python
df = pd.DataFrame(data)

df.head()
```

**Steg 5: Fixa kolumnnamnen**

1. Först hämtar vi dom...

```python
column_names = [column[0] for column in cursor.description]

column_names
```

2. Sen sätter vi dom

```python
df.columns = column_names

df.head()
```

**Steg 6: Spara data som CSV**

```python
df.to_csv("customers.csv",index=False)
```
